```python
# A. Analyze the performance of sales and revenue at the city and branch level
# Group by city and branch, calculate total sales and revenue
city_branch_sales = df.groupby(['City', 'Branch'])['Quantity'].sum()
city_branch_revenue = df.groupby(['City', 'Branch'])['Unit price'].sum()
```

```python
# Create bar charts
city_branch_sales.plot(kind='bar', title='Total Sales by City and Branch')
plt.xlabel('City and Branch')
plt.ylabel('Total Sales')
plt.show()

city_branch_revenue.plot(kind='bar', title='Total Revenue by City and Branch')
plt.xlabel('City and Branch')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# A. Analyze the performance of sales and revenue at the city and branch level
sales_by_city_branch = df.groupby(['City', 'Branch']).agg({'Quantity': 'sum', 'Unit price': 'sum'})
print("Performance of sales and revenue at the city and branch level:")
print(sales_by_city_branch)
```

```python
# Plotting sales and revenue at city and branch level
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Quantity', hue='Branch', data=sales_by_city_branch, palette='Set1')
plt.title('Total Sales at City and Branch Level')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.show()
```

```python
average_price_by_branch = df.groupby('Branch')['Unit price'].mean()
print(average_price_by_branch)
```

```python
# B. Average price of an item sold at each branch of the city
avg_price_by_branch_city = df.groupby(['City', 'Branch'])['Unit price'].mean()
print("\nAverage price of an item sold at each branch of the city:")
print(avg_price_by_branch_city)
```

```python
# Visualize average price per item by city and branch
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Unit price', hue='Branch', data=df)
plt.title('Average Price per Item by City and Branch')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.show()
```

```python
#C. Analyze the performance of sales and revenue, Month over Month across the Product line, Gender, and Payment Method, and identify the focus areas to get better sales for April 2019.

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Extract month and year
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
# Group by month, product line, gender, and payment method
monthly_sales = df.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment'])['Quantity'].sum()
monthly_revenue = df.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment'])['Unit price'].sum()
```

```python
# Create line plots with X-axis labels
fig, axes = plt.subplots(2, 1, figsize=(16, 14))

monthly_sales.unstack().plot(ax=axes[0], title='Monthly Sales by Product Line, Gender, and Payment Method')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Total Sales')
axes[0].legend(loc='upper right')
axes[0].tick_params(axis='x', rotation=45)


monthly_revenue.unstack().plot(ax=axes[1], title='Monthly Revenue by Product Line, Gender, and Payment Method')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Total Revenue')
axes[1].legend(loc='upper right')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(16, 14))

# Define titles
titles = ['Monthly Sales by Product Line, Gender, and Payment Method',
          'Monthly Revenue by Product Line, Gender, and Payment Method']

# Iterate over data and labels
for i, (data, ylabel, title) in enumerate(zip([monthly_sales, monthly_revenue], ['Total Sales', 'Total Revenue'], titles)):
    # Plot data
    data.unstack().plot(ax=axes[i], title=title)

    # Set labels and legend
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel(ylabel)
    axes[i].legend(loc='upper right')

    # Rotate x-axis labels
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
